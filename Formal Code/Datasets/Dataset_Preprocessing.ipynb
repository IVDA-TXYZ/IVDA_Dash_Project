### Data Cleaning

```python
print("Before drop operation: DF has {} rows. ".format(len(df)))
df = df.drop(df[df.popularity == 0].index)
print("Before drop operation: DF has {} rows. ".format(len(df)))
```

```python
print("Before drop_duplicates operation: DF has {} rows. ".format(len(df)))
df = df.drop_duplicates(keep='first')
print("Before drop_duplicates operation: DF has {} rows. ".format(len(df)))
```

## Genre Handling

```python
df = pd.read_csv('Clean_Dataset.csv')
```

```python
genre_list = df['genre'].apply(lambda x: x.split(', '))
```

```python
genres = []
for l in genre_list:
    # if set() in l:
    #     print(l)
    genres.extend(l)

genres = list(set(genres))
```

```python
# genres.remove('set()')
genres
```

```python
genres_dict = {x: '{}'.format(x) for x in genres}
genres_dict
```

```python
df
```

```python
for genre in genres:
    if genre == 'set()':
        continue
    df[genre] = 0

for i, _ in enumerate(genre_list):
    for genre in genre_list[i]:
        if genre == 'set()':
            continue
        df[genre][i] = 1
```
