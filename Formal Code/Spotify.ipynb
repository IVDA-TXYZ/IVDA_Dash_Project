```python
import re
#转换为列表
genre_list = list(data['genre'])
#去除“‘”
genre_list=str(genre_list).replace("'","").replace(" ","")
#重新转化为列表
genre_list_new=genre_list.strip('[')
genre_list_new=genre_list_new.strip(']')
genre_list_new=genre_list_new.split(',')
# print(genre_list_new)

count_list=dict()

for i in genre_list_new:
    if i in count_list:
        count_list[i]+=1
    else:
        count_list[i]=1
print(count_list)


```

```python
count_list = dict(sorted(count_list.items(), key=lambda x:x[1],reverse=True))

plt.figure(figsize=(6, 5)) 
plt.bar(count_list.keys(),count_list.values())
plt.xlabel('genre')
plt.xticks(rotation=270)
plt.ylabel('Counts of songs')
plt.title('Total songs based on genre')
```
